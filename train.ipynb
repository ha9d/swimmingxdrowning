```python
def calculate_iou(box1, box2):
    """
    Calculate the Intersection over Union (IoU) between two bounding boxes.

    Parameters:
        box1 (list): Bounding box coordinates [x_min, y_min, x_max, y_max].
        box2 (list): Bounding box coordinates [x_min, y_min, x_max, y_max].

    Returns:
        float: IoU score.
    """
    x1, y1, w1, h1 = box1
    x2, y2, w2, h2 = box2

    # Calculate intersection area
    intersection_area = max(0, min(x1 + w1, x2 + w2) - max(x1, x2)) * max(0, min(y1 + h1, y2 + h2) - max(y1, y2))

    # Calculate union area
    union_area = w1 * h1 + w2 * h2 - intersection_area

    # Calculate IoU
    iou = intersection_area / union_area if union_area > 0 else 0.0

    return iou
```

```python
import torch

def compute_accuracy(predictions, targets, iou_threshold=0.5):
    # Initialize counters for true positives, false positives, and false negatives
    tp = 0
    fp = 0
    fn = 0

    # Iterate over predictions and targets
    for pred, target in zip(predictions, targets):
        # Check if the prediction is empty (single value)
        if len(pred) == 1:
            # If the prediction is empty, consider it a false positive
            fp += 1
        else:
            # Extract bounding box coordinates and class from prediction
            pred_box = pred[:4]
            pred_class = int(pred[5])

            # Extract bounding box coordinates and class from target
            target_box = target[:4]
            target_class = int(target[5])

            # Calculate IoU between prediction and target
            iou = calculate_iou(pred_box, target_box)

            # Check if IoU is above the threshold
            if iou > iou_threshold and pred_class == target_class:
                tp += 1  # True positive
            else:
                fp += 1  # False positive

    # Calculate false negatives based on the total number of targets
    total_targets = len(targets)
    fn = total_targets - tp

    # Calculate accuracy
    accuracy = tp / (tp + fp + fn) if (tp + fp + fn) > 0 else 0.0

    return accuracy


predictions = torch.tensor([[0.1, 0.2, 0.4, 0.5, 0.9, 0]])
targets = torch.tensor([[0.1, 0.2, 0.4, 0.5, 0.9, 0]])

# Compute the accuracy
accuracy = compute_accuracy(predictions, targets)
print('Accuracy:', accuracy)
```
